```python
#checking for missing values 
for col in df.columns:
    missing_values = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, missing_values))
```

```python
df = df.replace({
    'location' : {
        ' ' : 'Africa'
    },
    'iso_code' : {
        'OWID_AFR' : 'AFR'

    }
})   
df
```

```python
#filling missing values 

df = df.fillna(method = 'bfill')
df = df.fillna({
 'continent' : 'Africa',
 'hosp_patients' : df['total_cases'],
 'icu_patients' : df['hosp_patients'],
 'icu_patients_per_million' : df['icu_patients'] * 0.0000001,
 'hosp_patients_per_million' : df['hosp_patients'] * 0.0000001,
 'weekly_icu_admissions' : df['icu_patients'] * 7,
 'weekly_hosp_admissions' : df['hosp_patients'] * 7, 
 'weekly_hosp_admissions_per_million' : df['weekly_hosp_admissions'] * 0.0000001,
 'weekly_icu_admissions_per_million' : df['weekly_icu_admissions'] * 0.0000001
  }) 
df = df.fillna(method = 'ffill')
df
```

```python
missing = np.sum(df.isnull())
missing
```

```python
df.shape
```

```python
#file sorted by date
df = df.sort_values(by = ['date'])
df
```

```python
#plot of new cases 
plt.style.use('fivethirtyeight')
plt.plot(df['total_cases'], df['new_cases'], 'k-',linewidth = 0.5,  label = 'new_cases')
plt.title('New_cases against total_cases')
plt.ylabel('New_cases')
plt.xlabel('Total_cases')
plt.grid(True)
plt.legend()
plt.show()
```

```python
plt.style.use('fivethirtyeight')
plt.plot(df['total_deaths'], df['new_deaths'], 'k-',linewidth = 0.5, label = 'new_deths')
plt.title('New_deaths against total_deaths')
plt.ylabel('New_deaths')
plt.xlabel('Total_deaths')
plt.grid(True)
plt.legend()
plt.show()
```
